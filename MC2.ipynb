## Calculate mean distances in n-dimensional cubes using random number simulation

```python
import numpy as np
```

```python
# Sample size
SampleSize = 10000000
# Run for dimension 1 ... max dimension
MaxDimension = 2

Dimensions = np.arange(1, MaxDimension+1, 1)
```

```python
# Run the calculations; reuse all previous squares for next dimension
SampleDistancesSquared = [0.0]*(SampleSize)
MeanDistances = []
StandardDeviations = []
for Index in range(MaxDimension):
    x1 = np.random.rand(SampleSize)
    x2 = np.random.rand(SampleSize)
    SampleDistancesSquared = SampleDistancesSquared + (x2-x1)**2
    SampleDistances = np.sqrt(SampleDistancesSquared)
    MeanDist = np.mean(SampleDistances)
    MeanDistances.append(MeanDist)
    StdDev = np.std(SampleDistances)
    StandardDeviations.append(StdDev)
```

Numerical observation is that limit of mean distance for large dimension is sqrt(dim/6).
Proof done.
Why does the standard deviation seem to converge to a constant value and does not scale with dimension?

```python
# demonstrate scaling behavior of mean distance with square root of the dimension
Factors = [0.0]*(MaxDimension)
Predictions = [0.0]*(MaxDimension)
for Index in range(MaxDimension):
    Factors[Index] = MeanDistances[Index]**2 / (Index+1)
    Predictions[Index] = np.sqrt((Index+1)/6)
#    print(Factors[Index], 1/Factors[Index])
```

```python
import matplotlib.pyplot as plt
plt.plot(Dimensions, MeanDistances)
plt.show()
plt.plot(Dimensions, np.log(np.abs(np.subtract(Predictions, MeanDistances))))
plt.show()
plt.plot(Dimensions, StandardDeviations)
plt.show
```

```python
Binwidth = 0.01
Bins = np.arange(min(SampleDistances), max(SampleDistances) + Binwidth, Binwidth)
plt.hist(SampleDistances, bins = Bins)
plt.show()
```
